```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.core.operations import SymmOp
with MPRester("c5FIVeAZuDMNrwBx") as m:
    structure = m.get_structure_by_material_id("mp-29868")
    w = CifWriter(structure)
    w.write_file('mp29868.cif')

    #La première opération est une 6-bar-axis
    print("Rotoinversion d'ordre 6 sur Li(0,0,0.5), dont la matrice de transformation associée est la    suivante :")
    symmetry_1 = SymmOp.from_xyz_string("-x+y,-x,-z") 
    #On cherche la matrice associée à la transformation
    print(symmetry_1.rotation_matrix)
    print(" ")
    
    #La seconde opération est un plan mirroir   
    print("Plan mirroir sur Li(1,0,0.5), dont la matrice de transformation associée est la suivante :")
    symmetry_2 = SymmOp.from_xyz_string("x,y,-z")
    print(symmetry_2.affine_matrix)   
    print(" ")

    #La troisième opération est une rotation C3-axis
    print("Rotation d'ordre 3 sur C2(0.67,0.33,0.38), dont la matrice de transformation associée est la suivante :")
    symmetry_3 = SymmOp.from_xyz_string("-y,x-y,z")
    print(symmetry_3.affine_matrix)
    
    #Dans les fenêtres suivantes se trouvent une image du LiAuC2 avant transformation
    #ainsi que les images des transformations citées ci-dessus dans le même ordre
    
    #Note : 
    #par soucis de clareté, les atomes n'ont pas été affichés dans les trois dernières images
    
```
