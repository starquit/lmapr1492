```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.io.cif import CifWriter
with MPRester("c5FIVeAZuDMNrwBx") as m:
    structure = m.get_structure_by_material_id("mp-29868")
    w = CifWriter(structure)
    w.write_file('mp29868.cif')
    
    xrdc = XRDCalculator(wavelength ='CuKa', symprec=0, debye_waller_factors=None)
    xrdc.get_plot(structure)
    pattern = xrdc.get_pattern(structure, scaled=True, two_theta_range=(0, 90)) #pattern est un xrd du api
    
    print("Voici les 2 theta pour les trois premiers pics")
    print(pattern.x[0],pattern.x[1],pattern.x[2]) #c'est mon 2 theta, les trois premiers
    print(" ") 
    print(pattern.y[0],pattern.y[1],pattern.y[2]) #c'est l'amplitude
    print(" ")
    print((pattern.hkls)[0],(pattern.hkls)[1],(pattern.hkls)[2]) #c'est les indices hkl
    print(" ")
    print((pattern.d_hkls)[0],(pattern.d_hkls)[1],(pattern.d_hkls)[2]) #ce sont les distances 
    
    
```
